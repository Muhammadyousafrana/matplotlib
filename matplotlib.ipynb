```python
car_sales
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```

```python
hear_disease = pd.read_csv("heart-disease.csv")
over_50 = hear_disease[hear_disease["age"] > 50]
over_50
```

```python
%matplotlib inline
plt.style.use("seaborn-v0_8-whitegrid")
fig, (ax0, ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True)
scatter = ax0.scatter(x = over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter", alpha=0.80, edgecolors="magenta", s=50)
ax0.set(title="Heart Disease An Cholestrol Level", ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(y=over_50["chol"].mean(), linestyle="--", c="red")
ax0.set_xlim([50, 80])
scatter = ax1.scatter(x = over_50["age"], y=over_50["thalach"], c= over_50["target"], cmap="viridis", alpha=0.80, edgecolors="magenta", s=50)
ax1.set(title="Heart Disease And Max Heart Rate Level", xlabel="age", ylabel="Thalach")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");
fig.suptitle("Heart Disease Analysis", weight="bold", c="red")
```
